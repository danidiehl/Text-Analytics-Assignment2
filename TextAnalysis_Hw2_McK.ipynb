```python
#formula to replace model names with brand names
#split each message into words. if word in search values, replace with matching idx number?
import numpy as np
def replace_model(s):
    mess=s.split()
    mess_rep=[]
    for word in mess:
        if word in model_list['Search1'].values:
            mask=(model_list['Search1']==word)
            word=model_list[mask]['Replace'].values[0]
            mess_rep.append(word)
        else:
            mess_rep.append(word)
    new_mess=' '.join(mess_rep)        
    return new_mess
    
            
replace_model('I love my legend') #YAY
replace_model('I love my quattro')
```

```python
#mapping that function to forum df
forum_data['message_clean']=forum_data['message'].map(replace_model)
```
